```python
import requests
import json
import pandas as pd
import time
from config import username, api_key
```

```python
flightNumber = 'AAL1618'
```

```python
baseurl = f"http://{username}:{api_key}@flightxml.flightaware.com/json/FlightXML2/"

addurl = f"FlightInfo?ident={flightNumber}&howMany=15"

fxmlurl = baseurl + addurl
```

```python
response = requests.get(fxmlurl).json()
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
data = response["FlightInfoResult"]
df = pd.DataFrame.from_dict(data["flights"], orient='columns')
df.head(5)
```

```python
for index, row in df.iterrows():
    
    if row["actualarrivaltime"] > 0:
        actualarrivaltime = time.strftime("%a, %d %b %Y %H:%M:%S +0000", time.localtime(row["actualarrivaltime"]))
        df.at[index, 'actualarrivaltime2'] = actualarrivaltime
    else:
        df.at[index, 'actualarrivaltime2'] = ""
    if row["actualdeparturetime"] > 0:
        actualdeparturetime = time.strftime("%a, %d %b %Y %H:%M:%S +0000", time.localtime(row["actualdeparturetime"]))
        df.at[index, 'actualdeparturetime2'] = actualdeparturetime
    else:
        df.at[index, 'actualdeparturetime2'] = ""
    
df.head(5)
```

```python
data = response["FlightInfoResult"]
filter_df = df.filter(['actualarrivaltime', 'actualdeparturetime', 'estimatedarrivaltime','filed_departuretime'] ,axis=1)

filter_df.head()
```

```python
arrival_status=filter_df['actualarrivaltime'] - filter_df['estimatedarrivaltime']
print(arrival_status);
```

```python
departure_status = filter_df['actualdeparturetime'] - filter_df['filed_departuretime']
print(departure_status);
```

```python
for index, row in df.iterrows():
    df.at[index, 'arrivalCount'] = arrival_status.iloc[index]
    df.at[index, 'departureCount'] = departure_status.iloc[index]
df.head()
```

```python
for index, row in df.iterrows():
    if row["arrivalCount"] > 0:
        #actualarrivaltime = time.strftime("%a, %d %b %Y %H:%M:%S +0000", time.localtime(row["actualarrivaltime"]))
        df.at[index, 'Flight_arrivalStatus'] = "Late"
    if row["arrivalCount"] < 0:
       # actualdeparturetime = time.strftime("%a, %d %b %Y %H:%M:%S +0000", time.localtime(row["actualdeparturetime"]))
        df.at[index, 'Flight_arrivalStatus'] = "Early"
    else:
        df.at[index, 'Flight_arrivalStatus'] = "On-Time"
    
df.head(5)
```
